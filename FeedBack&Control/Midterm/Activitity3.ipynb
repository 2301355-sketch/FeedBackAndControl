```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the transfer function coefficients
# H(s) = 3 / (8s^4 + 12s^3 + 26s^2 + 18s)
num = [3]
den = [8, 12, 26, 18, 0]
H = ct.TransferFunction(num, den)

# 2. Calculate the step response
# System contains a pole at the origin; response expected to grow.
t = np.linspace(0, 50, 1000)
time, amplitude = ct.step_response(H, T=t)

# 3. Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Subplot 1: Step Response
ax1.plot(time, amplitude, color='#2563eb', linewidth=2)
ax1.set_title('System Step Response', fontsize=14, fontweight='bold')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Amplitude')
ax1.grid(True, linestyle='--', alpha=0.7)

# Subplot 2: Pole-Zero Map
plt.sca(ax2)
ct.pzmap(H, plot=True)
ax2.set_title('System Pole-Zero Map', fontsize=14, fontweight='bold')
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
