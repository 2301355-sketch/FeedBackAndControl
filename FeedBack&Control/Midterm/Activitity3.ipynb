```python
import numpy as np
import control as ct
import matplotlib.pyplot as plt

# Define the polynomials
# Term A: (4s^2 + 4s + 6)
poly_a = np.poly1d([4, 4, 6])
# Term B: (4s^2 + 2s + 6)
poly_b = np.poly1d([4, 2, 6])

# Denominator calculation: (poly_a * poly_b) - 4s^2 - 36
den_poly = (poly_a * poly_b) - np.poly1d([4, 0, 0]) - np.poly1d([36])
num_poly = poly_b

# Create Transfer Function
sys = ct.TransferFunction(num_poly.coeffs, den_poly.coeffs)

print("Simplified Transfer Function G(s):")
print(sys)

# 1. Step Response
t, y = ct.step_response(sys)

# 2. Plotting
plt.figure(figsize=(12, 5))

# Step Response Plot
plt.subplot(1, 2, 1)
plt.plot(t, y, label='X2(t)', color='blue')
plt.title('Step Response of X2(s)')
plt.xlabel('Time (s)')
plt.ylabel('Displacement')
plt.grid(True)

# Pole-Zero Map
plt.subplot(1, 2, 2)
ct.pzmap(sys, plot=True)
plt.title('Pole-Zero Map')
plt.grid(True)

plt.tight_layout()
plt.show()
```
