```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np

# Define the Transfer Function: G(s) = 1 / (s^4 + 2s^3 + 3s^2 + 2s)
num = [1]
den = [1, 2, 3, 2, 0]
system = ct.TransferFunction(num, den)

# 1. Calculate Step Response with Amplitude 35
time = np.linspace(0, 50, 1000)
t, y = ct.step_response(system, T=time)
amplitude = 35
y_scaled = y * amplitude

# 2. Plotting Results
plt.figure(figsize=(14, 6))

# Subplot 1: Step Response
plt.subplot(1, 2, 1)
plt.plot(t, y_scaled, label=f'Step Amplitude = {amplitude}', color='b')
plt.title('System Step Response')
plt.xlabel('Time (seconds)')
plt.ylabel('Output Magnitude')
plt.grid(True, linestyle='--')
plt.legend()

# Subplot 2: Pole-Zero Map
plt.subplot(1, 2, 2)
ct.pzmap(system, plot=True, title='Pole-Zero Map')
plt.grid(True, linestyle='--')

plt.tight_layout()
plt.show()

# Output System Details
print("Derived Transfer Function G(s):")
print(system)
print("\nSystem Poles:")
print(system.poles())
```
