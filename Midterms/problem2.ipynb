```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficients of the transfer function
# Numerator: 1
# Denominator: s^4 + 2s^3 + 3s^2 + 2s + 0
num = [1]
den = [1, 2, 3, 2, 0]

# 1. Create and Print the Transfer Function
G = ctrl.TransferFunction(num, den)
print("Transfer Function:")
print(f"G(s) = {G}")
```

```python
# 2. Calculate and Print Poles and Zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
# 3. Generate Step Response Plot
# Compute the response to a unit step input
t, y = ctrl.step_response(G)

plt.figure(figsize=(10, 5))
plt.plot(t, y, color='#0d9488', linewidth=2)
plt.title('System Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid(True, alpha=0.3)
plt.show()
```
